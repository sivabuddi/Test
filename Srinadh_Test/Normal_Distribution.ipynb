```
import matplotlib.pyplot as plt
import scipy.stats
import numpy as np


x_min = 150
x_max = 190

mean = 170
std = 5

x = np.linspace(x_min, x_max, 100)

y = scipy.stats.norm.pdf(x,mean,std)

plt.plot(x,y, color='blue')

plt.grid()

plt.xlim(x_min,x_max)
plt.ylim(0,0.08)

plt.title('Normal distribution using scipy package and matplotlib with mean 170 and std 5',fontsize=15)

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```
